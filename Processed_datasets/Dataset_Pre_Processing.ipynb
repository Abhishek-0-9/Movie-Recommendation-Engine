```python
movies['genres'] = movies['genres'].apply(lambda x: x.split(', '))
movies['keywords'] = movies['keywords'].apply(lambda x: x.split(', '))
movies['ten_cast'] = movies['cast'].apply(lambda x: x.split(', ')[:10])
movies['cast'] = movies['cast'].apply(lambda x: x.split(', '))
movies['director'] = movies['director'].apply(lambda x: x.split(', '))
movies['production_countries'] = movies['production_countries'].apply(lambda x: x.split(', ')[:2])
movies['production_companies'] = movies['production_companies'].apply(lambda x: x.split(', ')[:3])

movies.sample(5)
```

```python
movies.info()
```

```python
movies.isna().sum()
```

```python
def merge(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(merge)
movies['ten_cast'] = movies['ten_cast'].apply(merge)
movies['director'] = movies['director'].apply(merge)
movies['genres'] = movies['genres'].apply(merge)
movies['keywords'] = movies['keywords'].apply(merge)
movies['production_countries'] = movies['production_countries'].apply(merge)
movies['production_companies'] = movies['production_companies'].apply(merge)
movies.sample(5)
```

```python
movies = movies.reset_index(drop=True)
movies.info()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['ten_cast']
```
