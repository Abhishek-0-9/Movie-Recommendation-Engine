#DataSets

```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
# https://www.kaggle.com/datasets/asaniczka/tmdb-movies-dataset-2023-930k-movies/data
movies = pd.read_csv("/content/drive/MyDrive/Data1.csv")

# https://www.kaggle.com/datasets/alanvourch/tmdb-movies-daily-updates/data
movies2 = pd.read_csv("/content/drive/MyDrive/Data2.csv")
```

```python
movies = pd.merge(movies, movies2, on=['id', 'title'])
```

```python
movies.info()
```

```python
movies = movies[['id', 'imdb_id_y', 'title', 'overview_y', 'genres_y', 'keywords', 'cast', 'director']]
movies = movies.rename(columns={
    'imdb_id_y': 'imdb_id',
    'overview_y': 'overview',
    'genres_y': 'genres'
})
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isna().sum()
```

```python
movies = movies.dropna()
movies.duplicated().sum()
```

```python
movies.drop_duplicates(inplace=True)
movies.duplicated().sum()
```

```python
movies.info()
```

```python
movies.isna().sum()
```

```python
# Convert 'cast', 'keywords', 'genres', and 'director' columns from comma-separated strings to lists
movies['genres'] = movies['genres'].apply(lambda x: x.split(', ') if pd.notna(x) else None)
movies['keywords'] = movies['keywords'].apply(lambda x: x.split(', ') if pd.notna(x) else None)
movies['cast'] = movies['cast'].apply(lambda x: x.split(', ')[:3] if pd.notna(x) else None)  # Take only first 3 cast members
movies['director'] = movies['director'].apply(lambda x: x.split(', ') if pd.notna(x) else None)

# Display the dataframe to check results
movies.sample(5)
```

```python
movies.info()
```

```python
movies.isna().sum()
```

```python
def merge(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(merge)
movies['director'] = movies['director'].apply(merge)
movies['genres'] = movies['genres'].apply(merge)
movies['keywords'] = movies['keywords'].apply(merge)
movies.sample(5)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

#Storyline

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']
```

```python
movies.head()
```

```python
storyline = movies[['id', 'imdb_id', 'title', 'tags']]
storyline['tags'] = storyline['tags'].apply(lambda x: " ".join(x))
```

```python
storyline['tags'][0]
```

```python
storyline['tags'] = storyline['tags'].apply(lambda x:x.lower())
```

```python
storyline['tags'][3]
```

```python
storyline['tags'][13]
```

```python
storyline.head()
```

```python
storyline['tags'] = storyline['tags'].apply(stem)
```

```python
storyline.sample(5)
```

```python
file_path = '/content/drive/My Drive/Storyline.csv'

# Save the DataFrame as a CSV file
storyline.to_csv(file_path, index=False)
```
